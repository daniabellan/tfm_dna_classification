```python
import torch
from torch.utils.data import ConcatDataset
import matplotlib.pyplot as plt
import os
import sys

os.chdir("/home/diz-wsl/personal_repos/tfm/")
print(os.getcwd())

from src.dataset.load_dataset import RealDataset
from src.utils.common import load_config

torch.manual_seed(42)
```

```python
def load_dataset(config: dict) -> ConcatDataset:
    """
    Loads multiple datasets from the specified paths and concatenates them into a single dataset.

    Args:
        config (dict): Configuration dictionary containing dataset paths.

    Returns:
        ConcatDataset: A PyTorch dataset containing all loaded datasets.
    """
    datasets = [
        RealDataset(config=config, dataset_path=dataset_path, class_idx=class_idx)
        for class_idx, dataset_path in enumerate(config["dataset_paths"])
    ]

    return ConcatDataset(datasets)
```

```python
# Load all configurations from the training experiment file
config = load_config("src/configs/experiment/experiment_1.yaml")

# Load and concatenate datasets
full_dataset = load_dataset(config["dataset"])
```

```python
w_idx = 2

plt.figure(figsize=(24, 15))

###############
# First dataset
###############
# First dataset without processing
first_dataset = full_dataset.datasets[0].data[0]
first_w_size = full_dataset.datasets[0].window_size

segment = [w_idx * first_w_size, ((w_idx * first_w_size)+first_w_size)]
plt.subplot(3, 2, 1)  # (Filas, Columnas, ndice)
plt.plot(first_dataset.signal_pa[segment[0]:segment[1]])
plt.title('First dataset signal without processing')
plt.xlabel(f'Time (window size {first_w_size})')
plt.ylabel('Signal value')
plt.grid(True)

# First dataset with processed data
plt.subplot(3, 2, 2)
plt.plot(first_dataset.full_processed_signal[segment[0]:segment[1]])
plt.title('First dataset signal processed')
plt.xlabel(f'Time (window size {first_w_size})')
plt.ylabel('Signal value')
plt.grid(True)


plt.tight_layout()  
plt.show()
```
